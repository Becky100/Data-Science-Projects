```python
import pandas as pd
```

```python
Raw_data=pd.read_csv(r'D:\\movies_metadata.csv',low_memory=False)
Raw_data.head(4)
```

```python
#cleaning the data
#check for null values in the title, vote_average and vote_count
Raw_data[['title', 'vote_average','vote_count']].isnull().sum()
```

```python
#drop rows with null values
clean_data=Raw_data.dropna(subset=['title','vote_average','vote_count'])
clean_data=clean_data.drop_duplicates()
clean_data.head()
```

```python
#calculate the weighted average 
#Formula (check website)
#calculating the mean vote across the whole report, C
C=clean_data['vote_average'].mean()
C
```

```python
M=clean_data.quantile(0.95)
M
```

```python
#Calculate M, the number of votes that a movie must have to be considered. let's take it as the 95% percentile
M=clean_data.quantile(0.95)
M=M['vote_count']
M
```

```python
#Filter out movies with a vote less than 434
#create a new dataframe without interferring with the original dataframe. Use the the copy method
cons_movies=clean_data.copy()
clean_data.equals(cons_movies)
```

```python
cons_movies.shape
```

```python
cons_movies['vote_count'].head(2)
```

```python

cons_movies=cons_movies.loc[cons_movies['vote_count'] >=M]
cons_movies.head()
```

```python
#calculate the weighted rating
#Define a function weighted_rating that returns the weighted rating
def weighted_rating(cons_movies,M=M,C=C):
    V=cons_movies['vote_count']
    R=cons_movies['vote_average']
    weightedrating=((V/(V+M))*R)+((M/(V+M))*C)
    return weightedrating
```

```python
#Defining a new feature called score. Applying the weighted rating function returns the 'score' for each movie
cons_movies['score'] = cons_movies.apply(weighted_rating, axis=1)
cons_movies['score'].head()
```

```python
#sort out the movies in descending order and output the top 10 movies
cons_movies=cons_movies.sort_values('score',ascending=False)
cons_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```
